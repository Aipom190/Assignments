```python
#標題、副標題、預覽、每章封面圖片連結、章節連結、小說內文
from tool import get_soup
import time
import json
from bs4 import BeautifulSoup
import pandas as pd
import os

#找頁碼
url = 'https://projectmoon.postype.com/series/491543/the-distortion-detective'
soup = get_soup(url)
pages = soup.find('ul', class_='pagination pagination').find_all('li')
page_number = pages[-2].find('span').text
time.sleep(3)

m = 1
novel_info = []

#從最後一頁開始
for n in range(int(page_number), 0, -1):
    page_url = f'https://projectmoon.postype.com/series/491543/page/{n}'
    page_soup = get_soup(page_url)
    chapters_list = page_soup.find_all('div', class_='component-post list') #這一頁的所有章節
    
    #因為要從網頁裡最下面的第一章開始抓，所以間隔取-1
    for k in range(len(chapters_list)-1, -1, -1):
        chapter_info = []
        try:
            chapter = chapters_list[k]
            contents = chapter.find('div', class_='post-data')

            #標題
            title = contents.find('h3').text
            if title.startswith('[Report]'): #排除人物介紹
                continue
            else:
                chapter_info.append(title)

            #副標題
            subtitle = contents.find('h4').text
            chapter_info.append(subtitle)

            #預覽
            preview = contents.find('p').text
            chapter_info.append(preview)

            #封面圖片連結
            cover = chapter.find('div', class_='post-data-thumbnail')
            coverpic = cover.find('img').get('src')
            chapter_info.append(coverpic)

            #章節連結
            link = contents.find('a').get('href')
            chapter_info.append(link)
            time.sleep(5)
            
            #小說內文
            print(f'第{m}章')
            link_number = link.strip('https://projectmoon.postype.com/post/')
            chapter_url = 'https://www.postype.com/api/post/content/' + link_number
            chapter_soup = get_soup(chapter_url)
            chapter_contents = chapter_soup.find('p').text
            contents_dict = json.loads(chapter_contents)
            chapter_text = contents_dict['data']['html']
            text = BeautifulSoup(chapter_text, 'lxml')
            text_list = text.find_all('p')
            for paragraph in text_list:
                if paragraph.text.startswith('Any subscriptions or financial support'):
                    continue
                elif paragraph.text.startswith(' Wanna support'):
                    continue
                elif paragraph.text.startswith('- What the next'):
                    break
                else:
                    print(paragraph.text)

            novel_info.append(chapter_info)
            m += 1
            time.sleep(6)

        except AttributeError: #最後一則貼文是漫畫
            continue

novel_path='Novel/Info/'
if not os.path.exists(novel_path):
    os.makedirs(novel_path)
    print(f'{novel_path}建立成功！')
else:
    print(f'{novel_path}已存在！')

info = pd.DataFrame(novel_info, columns=['Title', 'Subtitle', 'Preview', 'Cover', 'Link'])
info.to_csv(f'{novel_path}novel_info.csv', encoding='utf-8-sig')
print('novel_info.csv已儲存')
```

```python
#章節內的插圖
import requests
import time
import os

pic_url = 'https://libraryofruina.huijiwiki.com/wiki/%E6%89%AD%E6%9B%B2%E4%BE%A6%E6%8E%A2'
pic_soup = get_soup(pic_url)
pic_links = pic_soup.find_all('li', class_='gallerybox')
time.sleep(3)

#找出放各張圖片的網頁的網址，存在pics_list裡
pics_list=[]
for pic_link in pic_links:
    pics = 'https://libraryofruina.huijiwiki.com/'+pic_link.a.get('href')
    pics_list.append(pics)

#到放各張圖片的網頁裡抓圖片網址
pic_list = []
for pics in pics_list:
    soup = get_soup(pics)
    pic = soup.find('div', class_='fullImageLink').img.get('src')
    pic_list.append(pic)
    
pic_path='Novel/Gallery/'
if not os.path.exists(pic_path):
    os.makedirs(pic_path)
    print(f'{pic_path}建立成功！')
else:
    print(f'{pic_path}已存在！')

for i in range(0, len(pic_list)): 
    resp = requests.get(pic_list[i])
    with open(f'Novel/Gallery/{i}.jpg', 'wb') as f:
        f.write(resp.content)
        print(f'第{i}張圖片已儲存！')
```
